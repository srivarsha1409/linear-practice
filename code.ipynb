```python
#4
import pandas as pd
from sklearn.linear_model import LinearRegression


df = pd.read_csv("C:/AI WORKSHOP/aug 19 practice/house_price_multifeature_150.csv")


X = df[["size_m2", "bedrooms"]] 
y = df["price_k"]               

model = LinearRegression().fit(X, y)

print("Coefficient for size_m2 (β1):", model.coef_[0])
print("Coefficient for bedrooms (β2):", model.coef_[1])
print("Intercept (β0):", model.intercept_)
print("R²:", model.score(X, y))

# Prediction for size=85, bedrooms=3
pred = model.predict(pd.DataFrame([[85, 3]], columns=["size_m2", "bedrooms"]))[0]
print("Predicted price for size=85, bedrooms=3:", pred)
```

```python
#3
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


df = pd.read_csv("C:/AI WORKSHOP/aug 19 practice/salary_experience_150.csv")

# X = experience, y = salary
X = df[["experience_years"]]
y = df["salary_k"]


model = LinearRegression().fit(X, y)

# Predict salary at 7 years
pred_7 = model.predict(pd.DataFrame([[7]],columns=["experience_years"]))[0]
print("Predicted salary at 7 years:", pred_7)

plt.scatter(X, y, color="blue")
plt.plot(X, model.predict(X), color="orange")
plt.scatter(7, pred_7, color="red", s=100, marker="x")  
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```
